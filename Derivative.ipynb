**For Error Calculation**

```python
x=2
f_x1=x**2
h=0.00000002
f_x_h=(x+h)**2
der=(f_x_h-f_x1)/h
print(der)
error=abs(der-2*x)/4
print(error)
```

```python
x=np.pi/4
f_x1=np.cos(x)
h=1e-10
f_x_h=np.cos(x+h)
der=(f_x_h-f_x1)/h
print(der)
error=abs(der+np.sin(x))/np.sin(x)
print(error)
```
